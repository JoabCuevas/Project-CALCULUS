```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square
from scipy.integrate import quad
from math import*
from sympy import fourier_series, pi
from sympy.abc import x
```

```python
x=np.arange(-np.pi,np.pi,0.001) #x axis has been chosen from ‚ÄìœÄ to +œÄ, value 
#of 1 smallest square along x axis is 0.001 

y=square(x) #defining square wave function ùë¶ =‚àí1, ùëìùëúùëü ‚àí ùúã ‚â§ ùë• ‚â§ 0
#y= +1, ùëìùëúùëü 0 ‚â§ ùë• ‚â§ ùúã

#define fuction

fc=lambda x:square(x)*cos(i*x)  #i :dummy index

fs=lambda x:square(x)*sin(i*x)

n=50 #max value of I, not taken infinity, better result with high value

An=[] # defining array

Bn=[]

sum=0

for i in range(n):
    an=quad(fc,-np.pi,np.pi)[0]*(1.0/np.pi)

    An.append(an)

    
for i in range(n):

    bn=quad(fs,-np.pi,np.pi)[0]*(1.0/np.pi)

    Bn.append(bn) #putting value in array Bn

    
for i in range(n):

    if i==0.0:
        sum=sum+An[i]/2
    else:
        sum=sum+(An[i]*np.cos(i*x)+Bn[i]*np.sin(i*x))

plt.plot(x,sum,'g')

plt.plot(x,y,'r--')

plt.title("fourier series for square wave")

plt.show()
```
